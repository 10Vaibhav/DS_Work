# Expressing Conditional Logic as Array Operations

```python
# The numpy.where function is a vectorized version of the ternary expression "x if condition else y" 
# suppose we had a Boolean array and two arrays of values:
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
```

```python
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
cond = np.array([True, False, True, True, False])
```

```python
# The zip() function combines multiple iterables (like lists, arrays, etc.) into a single iterable of tuples,
# where each tuple contains the corresponding elements from the input iterables.
result = [(x if c else y) for x,y,c in zip(xarr, yarr, cond)]
```

```python
result
```

This has multiple problems. First, it will not be very fast for large arrays (because all the work is being done in interpreted Python code). Second, it will not work with multidimensional arrays.

```python
# with numpy.where you can do this with a single function call: 
result = np.where(cond, xarr, yarr)
```

```python
result
# The second and third arguments to numpy.where don’t need to be arrays; one or both of them can be scalars. 
```

A typical use of where in data analysis is to produce a new array of values based on another array.

```python
# Suppose you had a matrix of randomly generated data and you wanted to replace all positive values with 2 and all negative values with –2.
# This is possible to do with numpy.where:

arr = rng.standard_normal((4,4))
```

# Example: Random Walks

The Simulation of random walks provides an illustrative application of utilizing array operations.

```python
# Let's first consider a simple random walk starting at 0 with steps of 1 and -1 occurring  with equal probability.
# Here is a pure python way to implement a single random walk with 1,000 steps using the built-in random module: 

import random 
position = 0 

walk = [position]
nsteps = 1000

for _ in range(nsteps):
    step = 1 if random.randint(0, 1)else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

you might make the observation that walk is the cumulative sum of the random steps and could be evaluated as an array expression. 

Thus, i use the numpy.random module to draw 1,000 coin flips at once, set these to 1 and -1, compute the cumulative sum: 

```python
nsteps = 1000

rng = np.random.default_rng(seed = 12345) # fresh random generator
```

```python
draws = rng.integers(0,2, size = nsteps)
```

```python
steps = np.where(draws == 0,1, -1)
```

```python
walk = steps.cumsum()
```

From this we can begin to extract statistics like the minimum and maximum value along the walk’s trajectory:



```python
walk.min()
```

```python
walk.max()
```

# simulating Many Random Walks at once

```python
nwalks = 5000
```

```python
nsteps = 1000
```

```python
draws = rng.integers(0,2,size=(nwalks,nsteps))# 0 or 1 
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walks = steps.cumsum(axis = 1)
```

```python
walks
```

Now, we can compute the maximum and minimum values obtained over all of the walks:

```python
walks.max()
```

```python
walks.min()
```

Out of these walks, let’s compute the minimum crossing time to 30 or –30. This is slightly tricky because not all 5,000 of them reach 30. We can check this using the any method:

```python
hits30 = (np.abs(walks) >=30).any(axis = 1)
```

```python
hits30
```

```python
hits30.sum() # Number that hit 30 or -30 
```

We can use this Boolean array to select the rows of walks that actually cross the absolute 30 level, and call argmax across axis 1 to get the crossing times:

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis = 1)
```

```python
crossing_times
```

```python
crossing_times.mean()
```

Feel free to experiment with other distributions for the steps other than equal-sized coin flips. You need only use a different random generator method, like standard_normal to generate normally distributed steps with some mean and standard deviation:

```python
In [278]: draws = 0.25 * rng.standard_normal((nwalks, nsteps))
```

```python
draws
```

Keep in mind that this vectorized approach requires creating an array with nwalks * nsteps elements, which may use a large amount of memory for large simulations. If memory is more constrained, then a different approach will be required.
