Apply Custom Function through apply method and create custom function by lambda function

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else( 'Average' if x < 185 else 'Tall'))
```

```python
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'LightWeight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'MiddleWeight'
    else:
        return 'HeavyWeight'
```

```python
bios['Category'] = bios.apply(categorize_athlete, axis=1) # axis = 1 means rows & axis = 0 means Columns
```

# Advanced Functionality

.shift() .rank()  .rolling() .cumsum()

1. shift method

```python
# comparing revenue of today to yesterday : 
# positive value in shift means go forward and negative values means go backward.
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
coffee
```

```python
coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100 
```

```python
coffee
```

2. rank method

```python
bios.head()
```

```python
# comparing height of single athelet to all athelets and give its rank: 
bios['height_rank'] = bios['height_cm'].rank(ascending = False)
```

```python
bios.sort_values(['height_rank']).sample(10)[['name', 'height_rank']]
```

3. cumsum method

```python
# if you want to see the total revenue till current day
coffee['Cumulative_revenue'] = coffee['revenue'].cumsum()
```

```python
coffee
```

4. rolling method

```python
latte = coffee[coffee['Coffee Type'] == "Latte"].copy()
```

```python
latte
```

```python
latte["3day"] = latte['Units Sold'].rolling(3).sum()
```

```python
latte
```
